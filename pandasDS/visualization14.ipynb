```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data/space_missions.csv')
df.head()
```

```python
# always create a copy
space = df.copy()
space.head()
```

```python
# setting the 'Date' column as the index
space.set_index('Date', inplace=True)
```

```python
space.info()
```

```python
space['Cost'].head()
```

```python
space['Cost'].plot()
```

```python
space.plot()
```

**Bar Graphs**

```python
space['Company Name'].value_counts()
```

```python
space['Company Name'].value_counts().plot(kind='bar')
```

**Pie Charts**

```python
space['Status'].value_counts().plot(kind='pie', colors=['green', 'red'])
```

```python
space['Status'].value_counts().plot(kind='pie', explode=[0.1, 0], legend=True)
```

```python
plt.pie(space['Status'].value_counts())

plt.legend(space['Status'].value_counts().index)
```

```python
x = space['Company Name']
y = space['Cost']
plt.scatter(x, y, c=y, cmap='plasma', s=50, edgecolors='black')
plt.colorbar();
```
